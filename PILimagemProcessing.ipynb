## Beleza, vamos carregar uma imagem com PIL?
>ignore meu narcisismo..

```python
# drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# importe a biblioteca, please!
from PIL import Image
```

```python
image_davi_yale = "/content/drive/MyDrive/pukolabsAcademy_guides/imageProcessing/davi_yale.jpg"
```

```python
# mostre o tamanho da imagem
image = Image.open(image_davi_yale)
width, height = image.size
print("width:",str(width), "height:",str(height))
```

### **Resize() - Muito grande, vamos diminuir um pouco!**

```python
w = int(width * 0.8)
h = int(height * 0.8)

image = image.resize((w, h))
print("tamanho da imagem:", str(image.size))
image
```

### **Thumbnail() - outra forma de resized com a vantagem de redução proporcional.**

```python
# note que embora os parâmetros sejam 400, a proporção é 400 por 300!
image.thumbnail((400, 400))
image.save('davi_saara_thumbnail.jpg')
print("tamanho da imagem:", str(image.size))
image
```

## Antes de brincar, vamos conhecer a **alma de uma imagem?**

Você já percebeu que **imagem é uma função de manipulação de pixels**, geralmente representada por uma matriz onde em cada quadradinho pintamos uma cor, certo?

TOP! vamos ver isso então!

```python
# numpy nos ajuda com lista, matrizes, computação científica...
import numpy as np
```

```python
# mostre sua alma minha consagrada!
davi_saara_array = np.array(image)
davi_saara_array.shape
```

```python
# vamos carregar uma imagem básica preta para ver a alma dela.
black_basic_image = "/content/drive/MyDrive/pukolabsAcademy_guides/imageProcessing/black1.jpg"
```

```python
basic_image = Image.open(black_basic_image)
basic_image
```

```python
# mostre sua alma minha consagrada!
basic_image_array = np.array(basic_image)
basic_image_array.shape
```

**Vocês perceberam a diferença?** 
>(300, 400, 3) - uma matriz de 300 por 400, onde possui 3 canais de cores (RGB).

> (70, 68) - uma matriz de 70 por 68, onde há apenas 1 canal de cor.


### **Pegando valores de Pixel.**

```python
# vamos definir a coluna e linha que queremos pegar o valor de pixel.
# de davi_saara
davi_saara_array_pixel = davi_saara_array[200,200]
print(davi_saara_array_pixel)
# de black_basic_image
basic_image_array_pixel = basic_image_array[1,1]
print(basic_image_array_pixel)
```

## Vamos brincar um pouco!


### **Negative Inversion - subtraindo o RGB por 255.**

```python
from PIL import ImageOps

image_invert = ImageOps.invert(image)
print("ficou meio assustador, neh?")
image_invert
```

### **Mirror.**

```python
image_mirror = ImageOps.mirror(image)
image_mirror
```

### **Solarize - todos os pixels acima do nível de cinza são invertidos.**

```python
image_solarize = ImageOps.solarize(image)
image_solarize
```

## **Filtros**
> convolução e concatenação de matrizes para manipular o processamento da imagem, afinal imagem é função, lembra?

```python
# Blur
from PIL import ImageFilter 

image_blur = image.filter(ImageFilter.BLUR)
image_blur
```

```python
# Contour
image_contour = image.filter(ImageFilter.CONTOUR)
image_contour
```

```python
# Emboss
image_emboss = image.filter(ImageFilter.EMBOSS)
image_emboss
```

```python
# Detail
image_detail = image.filter(ImageFilter.DETAIL)
image_detail
```

```python
# edges
image_edges = image.filter(ImageFilter.FIND_EDGES)
image_edges
```

```python
# smooth_more
image_smooth_more = image.filter(ImageFilter.SMOOTH_MORE)
image_smooth_more
```

## Manipulando imagens com MatplotLib!

### **Plotando a imagem, ou sejam essa matriz ai de cima!**

```python
imgplot = plt.imshow(image)
```

### **Escala de cor.**
> se você parsear a matriz acima em um visualizador RGB, voce verá cada cor de cada pixel, vamos automatizar isso com matplotlib!

```python
imgplot = plt.imshow(image)
plt.colorbar()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(image)
ax.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(image)
imgplot.set_clim(0.0, 0.7)
ax.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```

## **Histogramas**

```python
# em relação ao valor do nível de cinza.
x=img_david_matplot[:,:,0]
# x coordenada. 
plt.xlabel("Value")
# y coordenada.
plt.ylabel("pixels Frequency")
# imshow com comparação do valor do nível de cinza..
plt.imshow(x,cmap="gray")
# plot 
plt.show()
plt.title("Histogramm")
plt.xlabel("Value")
plt.ylabel("pixels Frequency")
# plot histograma
plt.hist(x)
```
